```python
import pandas as pd
import numpy as np
import nbimporter
import random
import json

from feedback import wordle_feedback

pd.set_option('display.float_format', '{:.6f}'.format)
```

```python
words = pd.read_csv('data/unigram_freq.csv')
words_5chars = words[words['word'].str.len() == 5].sort_values('count', ascending=False).reset_index(drop=True)
words_5chars_filtered = list(words_5chars[:2600]['word'])
words_5chars_filtered
```

```python
def initial_entropy_calculation(guess, possible_words):
    # Dictionary to store how often each feedback pattern appears
    feedback_patterns = {}
    
    # Iterate over all possible words and calculate feedback for each
    for word in possible_words:
        feedback = tuple(wordle_feedback(word, guess))  # Tuple to make feedback hashable
        if feedback not in feedback_patterns:
            feedback_patterns[feedback] = 1
        else:
            feedback_patterns[feedback] += 1
    
    # Calculate entropy and track pattern contributions
    total_possible = len(possible_words)
    entropy = 0
    pattern_contributions = {}

    for feedback, count in feedback_patterns.items():
        prob = count / total_possible
        entropy_contribution = -prob * np.log2(prob)
        entropy += entropy_contribution
        # Store the contribution of each feedback pattern
        pattern_contributions[feedback] = entropy_contribution
    
    return entropy, pattern_contributions

initial_entropies = {}
counter = 1

for word in words_5chars_filtered:
    word_result = initial_entropy_calculation(word, words_5chars_filtered)
    initial_entropies.update({word: word_result[0]})
    print(f"Progress: {counter}/{len(words_5chars_filtered)}")
    counter+=1

initial_entropies = dict(sorted(initial_entropies.items(), key=lambda item: item[1], reverse=True))
```
